# Visualization 2 (Seaborn)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("cars.csv")
df.head(10)
```

## 1) Line plot

```python
# sns.lineplot?
```

```python
x = np.arange(10)
y = np.sqrt(x)

plt.figure(figsize=(6,6))
sns.lineplot(data = df,x = df.symboling,y = df.width)
# plt.plot(df['symbolinf',df['width']])
plt.show()
```

## 2) Histogram (Distplot)

```python
plt.figure(figsize=(6,5))
sns.distplot(df['price'])
plt.show()
```

### Displot()

```python
sns.displot(df.price,kde=True)
```

```python
sns.histplot(df.price,kde=True)
```

```python
width_mean = df["width"].mean()
width_median = df["width"].median()

plt.figure(figsize=(6,5))
sns.distplot(df["width"],hist=False)

plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="yellow")
plt.show()
```

## 3) Count plot

In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. 
You can use the hue parameter when creating histograms in seaborn to color the bars in the histogram based on the value of some specific variable. For example, adding to the lmplot function tells seaborn to colour the data points for smoker and non-smoker differently.

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style",hue='fuel-type') # hue = 'fuel-type' -> add this parameter for color encoding 
# df['body-style'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, y="make")
# plt.xticks(rotation=90)
plt.show()
```

```python
# Adding hue This parameter take column name for colour encoding
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style",hue="fuel-type")
plt.show()
```

## 4) Heatmap

```python
df_num = df.select_dtypes(include=['float64', 'int64','int32'])
```

```python
df_num.head()
```

```python
corr = df_num.corr()  # numeric variable
```

```python
corr
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True) #If annot is set to True, the text will be written on each cell.
plt.show()
```

- if the correlation is 1 then it is known as full correlation.
- if the correlation value is 0 to 1 this is known as positive correlation.
- if the correlation value is less than 0 means 0 to -1 this is known as negative correlation.


```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True,cmap="viridis")
plt.show()
```
